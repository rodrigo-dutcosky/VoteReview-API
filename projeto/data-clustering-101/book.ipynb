# | PROPOSTA

    1) Analisar os dados cadastrais & transacionais dos clientes - utilizar alguma "regra de negocio" 
    como premissa para classificar os clientes com base em seus habitos de consumo.
    
    2) Preparar a base de dados para utiliza-los como parametros de entrada do modelo de clusterizacao.
    
    3) Interpretar os resultados do modelo (classificacoes) - tentar encontrar padroes de semelhanca bem
    definidos entre os pontos de dados intra-clusters.
    
    

```python
### importacao dos dados para o ambiente de trabalho

import pandas as pd

df = pd.read_csv("cliente.csv")
trx = pd.read_csv("eventos.csv")
```

---

```python
### classificacao dos clientes com base no seu habito de consumo utilizando uma regra de negocio propria

### 1. calculando o volume de compras de cada cliente
volume = trx.groupby("id_cliente")["id"].count().reset_index()
volume.columns = ["id", "volume"]

### 2. calculando a receita gerada de cada cliente
receita = trx.groupby("id_cliente")["valor_total"].sum().reset_index()
receita.columns = ["id", "receita"]

### 3. agregando as features criadas nos passos 1 & 2 ao dataset de dados cadastrais dos clientes
df = df.merge(volume, on = "id", how = "left")
df = df.merge(receita, on = "id", how = "left")


### 4. dividindo a "receita" pelo "volume" de transacoes para encontrar o "ticket" medio de cada cliente
df["ticket"] = df["receita"] / df["volume"]

### 5. classificando os clientes como "gastador"; onde 1 = "sim, é gastador"; 0 = "nao, é meio mao de vaca"
df["gastador"] = df["ticket"].apply(lambda x: 1 if x > 1000 else 0)

### 6. removendo features que nao sejam mais necessarias
df.drop(["volume", "receita", "id", "ticket"], axis = 1, inplace = True)

```

---

```python
### dicionarios de referencia para os processos de "label encoding"

### 1. "ensino" & "status" ja estao em formato categorico

ENSINO = {
    "basico": 0, 
    "superior": 1, 
    "mestrado": 2, 
    "doutorado": 3
}
STATUS = {
    "casado": 0,
    "divorciado": 1,
    "solteiro": 2,
    "uniao_estavel": 3,
    "viuvo": 4
}

df["ensino"] = df["ensino"].apply(lambda x: ENSINO.get(x))
df["status"] = df["status"].apply(lambda x: STATUS.get(x))

### 2. "idade" foi sintetizada em categorias de faixa etaria previo ao label encoding

IDADE = {
    "< 20 anos": 0,
    "21 a 30 anos": 1,
    "31 a 40 anos": 2,
    "41 a 50 anos": 3, 
    "51 a 60 anos": 4,
    "> 60 anos": 5
}

df["idade"] = pd.cut(
    x = df["idade"], 
    bins = [-1, 20, 30, 40, 50, 60, 1000], 
    labels = IDADE.keys()
)
df["idade"] = df["idade"].apply(lambda x: IDADE.get(x))
```

---

```python
### treinando o modelo de clusterizacao K-Means()

from sklearn.cluster import KMeans

### 1. instanciando um objeto da classe KMeans passando o valor do hiperparametro "quantidade de clusters" = 4
model = KMeans(n_clusters = 4)

### 2. iniciando o treinamento do modelo com a funcao .fit()
model.fit(X = df[["ensino", "status", "filhos", "idade"]])

### 3. numeracao do cluster atribuido para cada ponto de dado pode ser acessada pelo atributo .labels_
df["cluster"] = model.labels_
```
